```python
## 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Derived class for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Derived class for Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

if __name__ == "__main__":
    # Create a Circle object
    circle = Circle(5)
    print(f"Area of the circle: {circle.area():.2f}")

    # Create a Rectangle object
    rectangle = Rectangle(4, 6)
    print(f"Area of the rectangle: {rectangle.area()}")
```

```python
## 3.  Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def display_type(self):
        print(f"Vehicle Type: {self.vehicle_type}")

# Derived class for Car
class Car(Vehicle):
    def __init__(self, vehicle_type, model):
        super().__init__(vehicle_type)
        self.model = model

    def display_model(self):
        print(f"Car Model: {self.model}")

# Further derived class for ElectricCar
class ElectricCar(Car):
    def __init__(self, vehicle_type, model, battery):
        super().__init__(vehicle_type, model)  # Call the constructor of Car
        self.battery = battery

    def display_battery(self):
        print(f"Battery Capacity: {self.battery} kWh")

if __name__ == "__main__":
    # Create an ElectricCar object
    my_electric_car = ElectricCar("Electric Vehicle", "Tesla Model S", 100)

    my_electric_car.display_type()
    my_electric_car.display_model()
    my_electric_car.display_battery()
```

```python
## 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number
        self.__balance = initial_balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ₹{amount}. New balance: ₹{self.__balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}. New balance: ₹{self.__balance}")
        else:
            print("Insufficient funds or invalid withdrawal amount.")

    def get_balance(self):
        return self.__balance

    def get_account_number(self):
        return self.__account_number
# Example usage:
account = BankAccount("1234567890", 1000)
print(f"Account Number: {account.get_account_number()}")
account.deposit(500)
account.withdraw(200)
print(f"Current balance: ₹{account.get_balance()}")
```

```python
## 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

class Instrument:
    def play(self):
        raise NotImplementedError("Subclasses must implement this method.")

# class for Guitar
class Guitar(Instrument):
    def play(self):
        return "Strumming the guitar!"

# class for Piano
class Piano(Instrument):
    def play(self):
        return "Playing the piano!"

def perform(instrument):
    print(instrument.play())

if __name__ == "__main__":

    my_guitar = Guitar()
    my_piano = Piano()

    perform(my_guitar)
    perform(my_piano)
```

```python
## 12.  Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)


if __name__ == "__main__":

    student1 = Student("Akash", [100, 90, 88, 92])
    student2 = Student("Aryan", [70, 75, 80, 75, 60])

    print(f"{student1.name}'s average grade: {student1.average_grade():.2f}")
    print(f"{student2.name}'s average grade: {student2.average_grade():.2f}")
```

```python
## 14.  Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate
    def calculate_salary(self, hours_worked):
        return hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus):
        super().__init__(name, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self, hours_worked):
        base_salary = super().calculate_salary(hours_worked)
        return base_salary + self.bonus

employee = Employee("Akash", 20.0)
employee_salary = employee.calculate_salary(40)
print(f"{employee.name}'s salary:₹{employee_salary}")

manager = Manager("Aryan", 30.0, 500.0)
manager_salary = manager.calculate_salary(40)
print(f"{manager.name}'s salary: ₹{manager_salary}")
```

```python
## 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo!"

class Sheep(Animal):
    def sound(self):
        return "Baa!"

cow = Cow()
sheep = Sheep()

print(f"The cow says: {cow.sound()}")
print(f"The sheep says: {sheep.sound()}")
```

```python
## 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def display_info(self):
        return f"House Address: {self.address}, Price: ₹{self.price:.2f}"

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def display_info(self):
        """Display information about the mansion, including the number of rooms."""
        base_info = super().display_info()
        return f"{base_info}, Number of Rooms: {self.number_of_rooms}"

if __name__ == "__main__":
    house = House("111 Bihar,India", 250000)
    print(house.display_info())
    mansion = Mansion("222 Dehlhu", 1500000, 5)
    print(mansion.display_info())
```
